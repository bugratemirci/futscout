```python
from keras.models import Sequential
import keras
from keras.layers import LSTM, Dense, Dropout

def createModel(X_train):
    regressor = Sequential()

    regressor.add(LSTM(units=100, return_sequences=True,
                       input_shape=(X_train.shape[1], 1)))
    regressor.add(Dropout(0.2))

    regressor.add(LSTM(units=100, return_sequences=True))
    regressor.add(Dropout(0.2))

    regressor.add(LSTM(units=100, return_sequences=True))
    regressor.add(Dropout(0.2))

    regressor.add(LSTM(units=100))
    regressor.add(Dropout(0.2))

    regressor.add(Dense(units=1))

    regressor.compile(optimizer="adam", loss='mse')

    return regressor
```

# Verilerin Modele Uygun Hale Getirilmesi

```python
total_season = 10
for i in range(7):
  X_train = [df.iloc[i, 1:total_season].values for i in range(800)]
  y_train = [df.iloc[i, total_season:total_season+1].values for i in range(800)]
  X_test = [df.iloc[i, 1:total_season].values for i in range(800,1000)]
  y_test = [df.iloc[i, total_season:total_season+1].values for i in range(800,1000)]

  y_train = np.array(y_train, dtype=np.float32).reshape(-1,1)
  y_test = np.array(y_test, dtype=np.float32).reshape(-1,1)
  
  mms = MinMaxScaler(feature_range=(1, 1.25))

  X_train = mms.fit_transform(X_train)
  X_test = mms.fit_transform(X_test)
  y_test = mms.fit_transform(y_test)
  y_train = mms.fit_transform(y_train)

  regressor = createModel(X_train)
  model_name = str(total_season - 1) + 'season/'
  model_path = 'models/goals/' + model_name
  model_checkpoint_callback = keras.callbacks.EarlyStopping(
          monitor="val_loss",
          min_delta=0,
          patience=6,
          verbose=1,
          mode="auto",
          baseline=None,
          restore_best_weights=False,
      )


  history = regressor.fit(X_train, y_train, epochs=100, batch_size=20,
                          verbose=1, validation_data=(X_test, y_test), callbacks=[model_checkpoint_callback])
  regressor.save(model_path + 'model.h5')

  plt.clf()

  plt.plot(history.history['loss'])
  plt.plot(history.history['val_loss'])
  plt.title('loss')
  plt.ylabel('loss')
  plt.xlabel('epoch')
  plt.legend(['train', 'val'], loc='upper left')
  plt.savefig(model_path + 'acc_loss_graph.png')
  total_season -= 1
```
