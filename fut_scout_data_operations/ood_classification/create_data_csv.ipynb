# Kaleci Verisi

```python
mevki = ["Goalkeeper" for i in range(1500)]
mevki = np.array(mevki).reshape(-1,1)
goalkeeper = np.concatenate((goalkeeper, mevki), axis = 1)
```

```python
p = [0.1428571428571429,0.1428571428571429,0.1428571428571429,0.1428571428571429,0.1428571428571429,0.1428571428571429,0.1428571428571429]

goalkeeper_clean_sheet = np.random.choice([40, 48, 49, 52, 54, 70, 95], p = p, size=(1,1500)).reshape(-1, 1)
goalkeeper_cs = np.concatenate((goalkeeper, goalkeeper_clean_sheet), axis=1)

accurate_long_ball = np.random.choice([25, 33, 45, 47, 65, 88, 96], p = p, size=(1,1500)).reshape(-1, 1)
goalkeeper_alb = np.concatenate((goalkeeper_cs, accurate_long_ball), axis=1)
side_ball = np.random.choice([40, 27, 35 , 52, 80, 89, 96], p = p, size=(1,1500)).reshape(-1, 1)
goalkeeper_sb = np.concatenate((goalkeeper_alb, side_ball), axis=1)
penalty_kick_off = np.random.choice([25, 33, 45, 67, 76, 80, 96], p = p, size=(1,1500)).reshape(-1, 1)
goalkeeper_ko = np.concatenate((goalkeeper_sb, penalty_kick_off), axis=1)
running_distance = np.random.choice([200, 250, 400, 450, 500, 300, 350], p = p, size=(1,1500)).reshape(-1, 1)
goalkeeper_rd = np.concatenate((goalkeeper_ko, running_distance), axis=1)
number_of_matches = np.random.choice([90, 102, 140, 145, 149, 150, 155], p = p, size=(1,1500)).reshape(-1, 1)

ind=np.argsort(goalkeeper_rd[:,3])
sorted_array_by_cs = goalkeeper_rd[ind]
ind2 = np.argsort(number_of_matches[:,0])
sorted_array_by_nom = number_of_matches[ind2]
goalkeeper_nom = np.concatenate((sorted_array_by_cs, sorted_array_by_nom), axis=1)
```

```python
index = [i for i in range(1500)]
np.random.shuffle(index)

goalkeeper_nom = goalkeeper_nom[index]

import pandas as pd
df = pd.DataFrame(goalkeeper_nom, columns=['Name', 'Age', 'Position', 'Clean Sheet', 'Long Ball', 'Side Ball', 'Penalty Kick Off', 'Running Distance', 'Number of Matches'])
df.head(1500)
```

## Kümeleme Kaleci

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))

df_age = df.iloc[:, 1:2]
df_other = df.iloc[:, 3:]
df_last = pd.concat([df_age, df_other], axis = 1)
df_last_val = df_last.drop(['Running Distance', 'Number of Matches', 'Age'], axis=1)

classes = AgglomerativeClustering(n_clusters=4).fit_predict(df_last_val)


goalkeeper_nom_last = np.concatenate((goalkeeper_nom, classes.reshape(-1,1)), axis = 1)
df = pd.DataFrame(goalkeeper_nom_last, columns=['Name', 'Age', 'Position', 'Clean Sheet', 'Long Ball', 'Side Ball', 'Penalty Kick Off', 'Running Distance', 'Number of Matches', 'Classes'])

boy = np.random.choice([175, 186, 182, 195, 198, 177, 190], p = p, size=(1,1500)).reshape(-1, 1)
kilo = np.random.choice([76, 82, 85, 80, 79, 77, 70], p = p, size=(1,1500)).reshape(-1, 1)
cols = ['Name', 'Position', 'Age','Height', 'Weight', 'Clean Sheet', 'Long Ball', 'Side Ball', 'Penalty Kick Off', 'Running Distance', 'Number of Matches', 'Classes']
df['Height'] = boy
df['Weight'] = kilo
df = df[cols]
df.to_csv('goalkeeper.csv')
df
```
