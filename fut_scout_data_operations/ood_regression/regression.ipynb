# Veri setinin çekilmesi

```python
train_data = np.load('/content/drive/MyDrive/proje_1_calismalar/kendi_urettigimiz_veri_regression/train/ds7.npy')
test_data = np.load('/content/drive/MyDrive/proje_1_calismalar/kendi_urettigimiz_veri_regression/test/ds7.npy')
```

# X ve Y ayrımının yapılması
- Train

```python
X_train = []
y_train = []


for i in range(6, len(train_data),7):
    X_train.append(train_data[i-6:i])
    y_train.append(train_data[i])

X_train = np.array(X_train)
y_train = np.array(y_train).reshape(-1,1)

mms = MinMaxScaler()
X_train = mms.fit_transform(X_train)
y_train = mms.fit_transform(y_train)
```

# X ve Y ayrımının yapılması 
- Test

```python
X_test = []
y_test = []


for i in range(6, len(test_data), 7):
    X_test.append(test_data[i-6:i])
    y_test.append(test_data[i])

X_test = np.array(X_test)
y_test = np.array(y_test).reshape(-1,1)

print(X_test)
print(y_test)

mms = MinMaxScaler()
X_test = mms.fit_transform(X_test)
y_test = mms.fit_transform(y_test)
```

# Modelin oluşturulması

```python
def createModel(X_train):
    regressor = Sequential()

    regressor.add(LSTM(units=100, return_sequences=True,
                       input_shape=(X_train.shape[1], 1)))
    regressor.add(Dropout(0.2))

    regressor.add(LSTM(units=100, return_sequences=True))
    regressor.add(Dropout(0.2))

    regressor.add(LSTM(units=100, return_sequences=True))
    regressor.add(Dropout(0.2))

    regressor.add(LSTM(units=100))
    regressor.add(Dropout(0.2))

    regressor.add(Dense(units=1))

    regressor.compile(optimizer="adam", loss='mse')

    return regressor
```

# Eğitim aşamasının başlangıcı

```python
model_checkpoint_callback = keras.callbacks.EarlyStopping(
        monitor="val_loss",
        min_delta=0,
        patience=3,
        verbose=0,
        mode="auto",
        baseline=None,
        restore_best_weights=False,
    )
model = createModel(X_train)

history = model.fit(X_train, y_train, epochs=100, batch_size=64,
                        verbose=1, validation_data=(X_test, y_test), callbacks=[model_checkpoint_callback])
model.save('/content/drive/MyDrive/proje_1/models/9_seasons.h5')

plt.clf()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.savefig('/content/drive/MyDrive/proje_1/figures/9_seasons.png')
```
